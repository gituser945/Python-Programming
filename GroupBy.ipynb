### Retrieve the group from group by object

```python
# importing dataset

fortune = pd.read_csv('datasets/fortune1000.csv',index_col="Rank")
sectors = fortune.groupby("Sector")
fortune.head()
```

```python
# let's retrieve the group of sector Energy

sectors.get_group("Energy")
```

```python
# sector max() function will return the maximum value from the left most column of dataframe
# notice that in company the last appereance of company name is appearing.

sectors.max().head(5)
```

```python
# min() function is reverse of the max() one

sectors.min().head()
```

```python
# let's check the sum() method it will only apply to numerical value
# Notice in output that it is applied to Revenue,profit and employees whic is numerical variable.

sectors.sum().head()
```

```python
# let's explore the mean method of the sector

sectors.mean().head()
```

### `Group_By` multiple columns

```python
# importing dataset and to apply group by to more columns we can pass multiple columns inside list of group by function

fortune = pd.read_csv('datasets/fortune1000.csv',index_col="Rank")
sectors_industry = fortune.groupby(["Sector","Industry"])
fortune.head()
```

```python
# let's apply first method
# notice in output that we are getting two indexed first with sector then industry.

sectors_industry.size()
```

```python
# let's call sum function

sectors_industry.sum().head()
```

### The `.agg()` method

```python
# agg() function takes values as dictionary and it will take key as column name and value as function to perform on that column

sectors.agg({'Revenue' : 'sum',
            'Employees' : 'sum',
            'Profits' : 'mean'}).head()
```

```python
# if we want to apply multiple aggregation functions at the same time we can provide all the method into list

sectors.agg(["sum","mean","median"]).head()
```

```python
sectors.agg({'Revenue' : ['sum','mean'],
            'Employees' : 'sum',
            'Profits' : 'mean'}).head()
```

### Iterating through groups

```python
# Importing dataset again

fortune = pd.read_csv('datasets/fortune1000.csv',index_col="Rank")
sectors_industry = fortune.groupby(["Sector","Industry"])
fortune.head()
```

```python
# let's break the problem into chunks
# first create new dataframe and copy the columns from our fortune dataframe
# then we'll use iteration to fill all rows into new dataframe

df = pd.DataFrame(columns=fortune.columns)
df
```

```python
# let's write for loop to insert data

for sector,data in sectors:
    highest_revenue = data.nlargest(1,"Revenue")
    df = df.append(highest_revenue)
```

```python
df
```

```python
# let's do this for city level

cities = fortune.groupby("Location")
df = pd.DataFrame(columns = fortune.columns)
df
```

```python
# let's iterate through cities

for city,data in cities:
    city_revenue = data.nlargest(1,"Revenue")
    df = df.append(city_revenue)
```
