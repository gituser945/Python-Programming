### Visualization using Matplot lib

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from numpy.random import randn,randint,uniform,sample
```

```python
df = pd.DataFrame(randn(1000),index = pd.date_range('2019-01-01',periods=1000),columns=['value'])
ts = pd.Series(randn(1000),index = pd.date_range('2019-01-01',periods=1000))
df.head()
```

```python
# iris data set present inside seaborn library

iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
ax = iris.plot(figsize = (20,10),title = 'Iris Dataset')
ax.set_xlabel('X - Axis')
ax.set_ylabel('Y - Axis')
```

```python
df1 = iris.drop(['species'],axis = 1) # droping species column
```

```python
df1 = df1.iloc[0]
```

```python
df1.plot(kind = 'bar',figsize = (10,5))
df1.plot.bar()
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
titanic['pclass'].plot(kind='bar')
```

```python
df = pd.DataFrame(randn(10,3),columns=['a','b','c'])
df.head()
```

```python
df.plot.bar(figsize = (10,5))
```

```python
# to flip a graph

df.plot.barh(figsize = (10,5))
```

```python
# histogram

iris.plot.hist(figsize = (10,5))
```

```python
iris.plot(kind = 'hist',stacked = True,bins =100,figsize= (10,5))
```

```python
df = iris.drop(['species'],axis = 1)
```

```python
df.diff().hist(figsize = (10,5),alpha = 0.2,color = 'r')
```

```python
# scatterplot

iris.plot.scatter(x = 'sepal_length',y = 'sepal_width',figsize = (10,5))
```

```python
iris.plot.scatter(x = 'sepal_length',y = 'sepal_width',c = 'petal_length',figsize = (10,5))
```

```python
iris.plot.scatter(x = 'sepal_length',y = 'sepal_width',c = 'petal_length',figsize = (10,5),s = 190)
```

```python
iris.plot.hexbin(x = 'sepal_length',y = 'sepal_width',figsize = (10,5),gridsize=10)
```
