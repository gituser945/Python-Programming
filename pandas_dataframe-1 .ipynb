### The `.rank()` Methods

```python
# let's import dataset first

NBA = pd.read_csv("pandas/nba.csv")
NBA.head(3)
```

```python
# let's first remove all NaN values and replace all salary's NaN with 0

NBA.dropna(how = 'all',inplace = True)
NBA['Salary'] = NBA['Salary'].fillna(0).astype("int")
NBA.head(3)
```

```python
# Rank method is used to rank the columns
# let's take a example where we want to rank the players based on thier salaries.
# And by default the rank method is sorted in ascending.
# Assigning the Rank into New column 'Salary Rank'

NBA['Salary Rank'] = NBA['Salary'].rank(ascending=False).astype("int")
```

```python
NBA.head()
```

```python
# let's check the salary and their ranks

NBA.sort_values('Salary',ascending=False).head()
```
