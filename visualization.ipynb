```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from plotly.offline import iplot
import plotly as py
import plotly.tools as tls
import cufflinks as cf
#%matplotlib inline
```

```python
py.offline.init_notebook_mode(connected = True)
cf.go_offline()
```

```python
df = pd.DataFrame(np.random.rand(300,3),columns = ['a','b','c'])
```

```python
df
```

```python
df.iplot(x = 'a',y = 'b')
cf.set_config_file(theme = 'solar')
```

```python
# using mode parameters

cf.set_config_file(theme = 'solar')
df.iplot(x = 'a',y = 'b',mode = 'markers',color = 'blue')
```

```python
# loading titanic dataset

titanic = sns.load_dataset('titanic')
```

```python
titanic
```

```python
cf.set_config_file(theme = 'solar')
titanic.iplot(x = 'sex',y = 'survived',kind = 'bar')
```

```python
# relation between fare and age

cf.set_config_file(theme = 'solar')
titanic.iplot(y = 'age',x = 'fare',mode = 'markers',color = 'blue')
```

```python
# plotting boxplot

cf.set_config_file(theme = 'solar')
df.iplot(kind = 'box')
```
