```python
# let's convert the Start Date which is currently in object datatype into date time format.

employees['Start Date'] # notice that this column is in object form
```

```python
# let's convert the Last Login Time which is currently in object datatype into date time format.

employees['Last Login Time'] = pd.to_datetime(employees['Last Login Time'])
```

```python
employees.head(3)
```

```python
# Let's convert Serior Management column into Boolean as it contains only True,False

employees['Senior Management'] = employees['Senior Management'].astype("bool")
```

```python
# let's convert Gender into category as it contains only "Male" and "Female"

employees['Gender'] = employees['Gender'].astype("category")
```

```python
# let's check the info as we have convert 5 columns datatype

employees.info()

# Notive the size of file is also reducing as we are converting to thier specific datatype from 62.5KB to 49.1+ KB
```

### Filter a dataframe based on a condition

```python
# now let's import using parse_date parameter

employees = pd.read_csv('pandas datasets//employees.csv',parse_dates=['Start Date','Last Login Time'])
```

```python
employees["Senior Management"] = employees['Senior Management'].astype("bool")
employees['Gender'] = employees['Gender'].astype("category")
employees.head()
```

```python
# filter employees dataframe based on gender = Male

employees[employees['Gender'] == 'Male'].head()
```

```python
# filter employees dataframe based on team = finance

employees[employees['Team'] == 'Finance'].head(3)
```

```python
# filter employees dataframe based on team not in marketing

employees[employees['Team'] != 'Marketing'].head()
```

```python
# filter who's salary is greated and equal to 10000

employees[employees['Salary'] >= 10000].head()
```

### Filter with `more than one condition` 

### `AND` Condition

```python
employees = pd.read_csv('pandas datasets//employees.csv',parse_dates=['Start Date','Last Login Time'])
employees["Senior Management"] = employees['Senior Management'].astype("bool")
employees['Gender'] = employees['Gender'].astype("category")
employees.head()
```

```python
# Filter with value in Gender is male and team is marketing

mask1 = employees['Gender'] == 'Male' 
mask2 = employees['Team'] == 'Marketing'

employees[mask1 & mask2].head()
```

### `OR` Condition

```python
employees = pd.read_csv('pandas datasets//employees.csv',parse_dates=['Start Date','Last Login Time'])
employees["Senior Management"] = employees['Senior Management'].astype("bool")
employees['Gender'] = employees['Gender'].astype("category")
employees.head()
```

```python
# Filter either management is true or start date before 1 jan 1990

mask1 = employees['Senior Management']
mask2 = employees['Start Date'] < '1990-01-01'

employees[mask1 | mask2].head(10)
```

```python
mask1 = employees['First Name'] == 'Robert'
mask2 = employees['Team'] == 'Client Services'
mask3 = employees['Start Date'] > '2016-06-01'

employees[(mask1 & mask2) | mask3].head()
```

### The `.isin()` Method

```python
employees = pd.read_csv('pandas datasets//employees.csv',parse_dates=['Start Date','Last Login Time'])
employees["Senior Management"] = employees['Senior Management'].astype("bool")
employees['Gender'] = employees['Gender'].astype("category")
employees.head()
```

```python
# extract value where team is either sales,marketing and client services

employees[employees['Team'].isin(['Marketing','Client Services','Sales'])].head(3)
```

### The `.isnull()` and `.notnull()` method

```python
employees = pd.read_csv('pandas datasets//employees.csv',parse_dates=['Start Date','Last Login Time'])
employees["Senior Management"] = employees['Senior Management'].astype("bool")
employees['Gender'] = employees['Gender'].astype("category")
employees.head(3)
```

```python
# extract values where team column has null

employees[employees['Team'].isnull()].head(3)
```

```python
# extract values where team column has not null

employees[employees['Team'].notnull()].head()
```

### The `.between()` Method

```python
employees = pd.read_csv('pandas datasets//employees.csv',parse_dates=['Start Date','Last Login Time'])
employees["Senior Management"] = employees['Senior Management'].astype("bool")
employees['Gender'] = employees['Gender'].astype("category")
employees.head(3)
```

```python
# Extract the employees having salary between 60000 and 70000

employees[employees['Salary'].between(60000,70000)].head()
```

```python
# Extract the employees who got bonus between 6 and 8 percent

employees[employees['Bonus %'].between(6.0,8.0)].head()
```

```python
employees[employees['Start Date'].between('1990-01-01','1993-01-01')].head()
```

```python
# extract employees who's login time between 8:30-10:00 am morning

employees[employees['Last Login Time'].between("08:30AM","10:00AM")].head()
```

### The `.duplicate()` Method

```python
employees = pd.read_csv('pandas datasets//employees.csv',parse_dates=['Start Date','Last Login Time'])
employees["Senior Management"] = employees['Senior Management'].astype("bool")
employees['Gender'] = employees['Gender'].astype("category")
employees.sort_values('First Name',inplace=True)
employees.head(3)
```

```python
employees[employees['First Name'].duplicated()].head()
```

```python
# using keep = False parameter of duplicated mathod

mask = ~employees['First Name'].duplicated(keep = False)
```

```python
employees[mask]
```

```python
mask = ~employees['First Name'].duplicated()
```

```python
employees[mask]
```
