#### change index level name using `set_names` method

```python
# importing dataset

bigmac = pd.read_csv('datasets/bigmac.csv',parse_dates=['Date'],index_col=['Date','Country'])
bigmac.sort_index(inplace=True)
bigmac.head()
```

```python
# let's change the name

bigmac.index.set_names(["Day","Location"],inplace=True)
bigmac.head()
```

```python
# let's rechange the Day to Date

bigmac.index.set_names(names = "Date",level=0,inplace = True)
bigmac.head()
```

### The `swaplevel` Method

```python
# importing dataset

bigmac = pd.read_csv('datasets/bigmac.csv',parse_dates=['Date'],index_col=['Date','Country'])
bigmac.sort_index(inplace=True)
bigmac.head()
```

```python
# let's swap the level when we have only 2 levels then we don't need to specify extra parameter it basically swap the two
# Notice that in output we have Country at the first and Date at the second

bigmac.swaplevel()
```

```python
# But if we have multiple index then we can do this by using swaplevel by providing name of the index or number
# order doesn't matter

bigmac.swaplevel("Date","Country")
bigmac.swaplevel("Country","Date")
```

```python
# by using index number

bigmac.swaplevel(1,0)
bigmac.swaplevel(0,1)
```

### The `.stack()` Method
It takes the top column and moves the index to main index on the left

```python
# import dataset

world = pd.read_csv("datasets/worldstats.csv",index_col=["country","year"])
world.head(3)
```

```python
# let's call .stack() method

world.stack()
```

```python
world.stack().to_frame()
```

### The `.unstack()` method
It will convert rows to column

```python
# import dataset

world = pd.read_csv("datasets/worldstats.csv",index_col=["country","year"])
world.head(3)
```

```python
# let's stack first

s = world.stack()
s
```

```python
 # lets use unstack to move to original

s.unstack()
```

```python
# let's call unstack method once again

s.unstack().unstack()
```

```python
# now let's use unstack by using index number

world.stack()
world.unstack(0).head(10)
```

### The `pivot` Method

```python
# import dataset

sales = pd.read_csv('datasets/salesmen.csv',parse_dates=["Date"])
sales.head()
```

```python
# let's check the unique values of Salesmen

sales['Salesman'].value_counts()
```

```python
# let's convert this column into category

sales['Salesman'] = sales['Salesman'].astype("category")
```

```python
# let's call pivot method

sales.pivot(index = "Date",columns="Salesman",values="Revenue").head(10)
```

### The `pivot_table` Method

```python
# import new dataset

food = pd.read_csv('datasets/foods.csv')
food.head(10)
```

```python
# let's call pivot_table function

food.pivot_table(values = 'Spend',columns='Gender',aggfunc="mean")
```

```python
# let's call pivot_table function

food.pivot_table(values = 'Spend',columns='Gender',aggfunc="sum")
```

```python
# let's check by item

food.pivot_table(values = 'Spend',columns='Item',aggfunc="mean")
```

```python
# let's check by item and gender

food.pivot_table(values = 'Spend',columns=['Item','Gender'],aggfunc="mean") 
```

```python
# let's check by item and gender

food.pivot_table(values = 'Spend',index =['Item','Gender'],columns = "City",aggfunc="mean") 
```

### The `pd.melt` 
This is kind of reverse to the pivot table it is actually takes the pivot table values and return the tabular form of it.

```python
# let's import dataset

quarter = pd.read_csv('datasets/quarters.csv')
quarter
```

```python
pd.melt(quarter,id_vars="Salesman",var_name="Quarters",value_name="Revenue")
```
