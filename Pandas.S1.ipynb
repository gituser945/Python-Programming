```python
# by default the header argument is false and if we make it true then system will throw original data without converting
# into header from row

file = pd.read_csv('Desktop/Canvas/examples/ex1.csv',header=None)
```

```python
# providing custom header name

file = pd.read_csv('Desktop/Canvas/examples/ex1.csv',names=['col1','col2','col3','col4','col5'])
```

```python
# reading multi indexes file system

file = pd.read_csv('Desktop/Canvas/examples/csv_mindex.csv',index_col=['key1','key2'])
```

```python
# while data reading if we want to replace 0's present in dataset and we want to convert the same into NaN values we use
# na_values = ['Zero']

file = pd.read_csv('Desktop/Canvas/examples/ex5.csv',na_values=['one']) # making ones as NaN value
```

```python
file
```

```python
# to skip any rows from dataset we can use skiprow

file = pd.read_csv('Desktop/Canvas/examples/ex5.csv',skiprows=[0]) # skipping 0th row
```

```python
file
```
