## Seaborn Library

```python
# importing libraries

import pandas as pd
import numpy as np
import seaborn as sns
```

```python
import ssl
ssl._create_default_https_context = ssl._create_unverified_context
```

```python
# loading tips dataset

tips = sns.load_dataset('tips')
```

```python
tips
```

```python
# checking tips vs total bill

sns.relplot(x = 'tip',y = 'total_bill',data = tips)
```

```python
# grouping the relation based on smokers and non-smokers

sns.relplot(x = 'tip', y = 'total_bill', data = tips, hue = 'smoker')
```

```python
# grouping based on daywise

sns.relplot(x = 'tip', y = 'total_bill',data = tips,hue = 'day')
```

```python
# Further grouping the above graph by smoker variable

sns.relplot( x = 'tip',y = 'total_bill', data = tips,hue = 'day',style = 'smoker')
```

```python
# creating new dataframe

df = pd.DataFrame(np.random.rand(300,2),columns = ['a','b'])
```

```python
df
```

```python
# checking relationship between a and b

sns.relplot(x = 'a',y = 'b',data = df,kind = 'line')
```

```python
# loading iris dataset

iris = sns.load_dataset('iris')
```

```python
iris
```

```python
# checking relationship between sepal_length and sepal_width

sns.scatterplot(x = 'sepal_length',y = 'sepal_width',data = iris)
```

```python
# plotting categorical vs numerical 

sns.catplot(x = 'day',y = 'tip',data = tips)
```

```python
sns.catplot(x = 'smoker',y = 'tip',data = tips)
```

```python
# boxplot using sns

sns.catplot(x = 'day',y = 'total_bill',data = tips,kind = 'box',hue = 'sex')
```

```python
# generating random numbers

x = np.random.randn(10000)
```

```python
# checking distribution of x

sns.distplot(x)
```

```python
# joint plot

sns.jointplot(x = tips['tip'], y = tips['total_bill'])
```

```python
# pair plot

sns.pairplot(iris)
```

```python
# linear model plot

sns.lmplot(x = 'total_bill',y='tip',data = tips)
```
