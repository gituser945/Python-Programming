### Pandas Vizualization

```python
# importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# creating dataset for this task

np.random.seed(20)
df = pd.DataFrame(np.random.randn(1000,1),index = pd.date_range("2021-09-01",periods = 1000),columns = ['Data'])
```

```python
df
```

```python
# let's visualize in order to see the trends

df.plot(figsize = (10,5))
```

```python
df.cumsum().plot(figsize =(10,5))
```

```python
# loading iris data

iris = sns.load_dataset('iris')
```

```python
iris
```

```python
gf = iris.plot(figsize = (10,5),title = "Iris Dataset")
gf.set_xlabel("Index")
gf.set_ylabel("This is Y axis")
```

```python
# fetching one record from dataset

iris.iloc[0]
```

```python
iris.iloc[0][0:4]
```

```python
iris.iloc[0][0:4].plot(kind = 'bar',figsize = (10,5))
```

```python
# loading titanic dataset

titanic = sns.load_dataset('titanic')
```

```python
titanic
```

```python
# ploting class 

titanic['pclass'].plot(kind = 'bar',figsize=(10,5))
```

```python
df1 = pd.DataFrame(np.random.rand(10,4),columns = ['a','b','c','d'])
```

```python
df1
```

```python
df1.plot(kind = 'bar',figsize = (10,6))
```

```python
# rotating the graph

df1.plot(kind = 'barh',figsize = (10,6))
```

```python
# plotting histogram

iris.plot(kind = 'hist',figsize = (10,6))
```

```python
# plotting sepal length distribution

iris['sepal_length'].plot(kind = 'hist',figsize = (10,6))
```

```python
iris['sepal_length'].plot(kind = 'density',figsize = (10,5))
```

```python
iris.describe()
```

```python
iris.plot(kind = 'hist',stacked = True,bins=100,figsize=(10,6))
```

```python
iris.hist(figsize = (15,10),color = 'olive',grid = False)
```

```python
# checking relationship between two variables

iris.plot.scatter(x = 'sepal_width',y = 'petal_width',figsize = (10,6))
```

```python
iris.plot.scatter(x = 'sepal_width',y = 'petal_width',c = 'petal_length',figsize=(10,6))
```

```python
iris.plot.scatter(x = 'sepal_width',y = 'petal_width',c = 'petal_length',figsize=(10,6),label = 'Points',s=80)
```

```python
iris.plot.hexbin(x = 'sepal_length',y = 'sepal_width',label = 'Hexabin Graph',gridsize = 10,figsize = (10,7))
```

```python
df1.iloc[0].plot.pie(figsize = (10,6))
```

```python
# plotting scatter matrix

pd.plotting.scatter_matrix(iris,figsize = (10,6))
```
