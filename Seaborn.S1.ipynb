### Seaborn Visualization

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# set background by using style

sns.set(style = 'darkgrid')
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.relplot(x = 'total_bill',y = 'tip',data=tips)
```

```python
sns.relplot(x = 'total_bill',y = 'tip',data=tips,hue = 'smoker',style = 'time')
```

```python
fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
sns.relplot(x = 'timepoint',y = 'signal',data = fmri,kind = 'line',ci = 'sd')
```

```python
sns.relplot(x = 'timepoint',y = 'signal',data = fmri,kind = 'line',estimator = None)
```

```python
sns.relplot(x = 'timepoint',y = 'signal',data = fmri,kind = 'line',hue = 'subject')
```

```python
sns.relplot(x = 'timepoint',y = 'signal',data = fmri,kind = 'line',hue = 'event',markers=True,dashes=False)
```

```python
sns.relplot(x = 'total_bill', y = 'tip',col = 'time',data = tips)
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
#catplot

sns.catplot(x = 'day',y = 'total_bill',data = tips)
```

```python
sns.catplot(x = 'day',y = 'total_bill',data = tips,jitter = False)
```

```python
sns.catplot(x = 'day',y = 'total_bill',data = tips,kind = 'swarm',hue = 'size')
```

```python
sns.catplot(x = 'day',y = 'tip',data = tips,kind = 'box',hue = 'sex')
```

```python
diamonds = sns.load_dataset('diamonds')
diamonds.head()
```

```python
sns.catplot(x = 'cut', y = 'price',kind = 'boxen',data = diamonds)
```

```python
from numpy.random import randn
import matplotlib.pyplot as plt
df = randn(1000)
```

```python
sns.distplot(df,kde = True ,hist = True,bins = 30,rug = False)
```

```python
iris = sns.load_dataset('iris')
sns.pairplot(iris)
```

```python
# Linear model plot

sns.lmplot(x = 'total_bill',y = 'tip',data = tips)
```
