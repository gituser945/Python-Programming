```python
# Concatenating and merging

df1 = pd.DataFrame({'A':[1,2,3],'B':[3,4,5],'C':[6,7,8]},index = [0,1,2])
df2 = pd.DataFrame({'A':[1,2,3],'B':[3,4,5],'C':[6,7,8]},index = [3,4,5])
df3 = pd.DataFrame({'A':[1,2,3],'B':[3,4,5],'C':[6,7,8]},index = [6,7,8])
```

```python
# concatenting using rows
df4 = pd.concat([df1,df2,df3],axis = 0)
```

```python
df4
```

```python
# concatenting using column
df5 = pd.concat([df1,df2,df3],axis = 1)
```

```python
df5
```

```python
# looking for column A

df5['A']
```

```python
# fill NA with 0

df5.fillna(value=0,inplace=True)
```

```python
df5
```

```python
# merging

left = pd.DataFrame({'Key':['k1','k2','k3'],'A':[1,2,3],'B':[4,5,6]})
right = pd.DataFrame({'Key':['k1','k2','k3'],'C':[1,2,3],'D':[4,5,6]})
```

```python
merge_data = pd.merge(left,right,how = 'inner',on = 'Key')
```

```python
merge_data
```
